```python
import random
```

```python
class Juego:

    def inicio_juego(self):
        comenzamos = input('Bienvenido al juego. ¿Qué te apetece jugar hoy?(Piedra papel o tijera, ahorcado o juego de preguntas)').lower()

        if comenzamos == 'piedra papel o tijera':
            self.piedra_papel_tijera()
        elif comenzamos == 'ahorcado':
            self.ahorcado()
        elif comenzamos == 'juego de preguntas':
            self.juego_preguntas_respuestas()
        elif comenzamos == 'no':
            print('Nos vemos pronto') 

    def piedra_papel_tijera(self):

        contador_jugador1 = 0
        contador_ordenador = 0

        # Creamos un bucle WHILE que irá repitiendo las jugadas hasta que alguno de los jugadores alcance la puntuación máxima (3 puntos)
        while contador_jugador1 < 3 and contador_ordenador < 3:
            eleccion_jugador1 = input('¿Qué opción eliges, piedra, papel o tijera').upper() 
            eleccion_ordenador = "".join(random.sample(['PIEDRA' , 'PAPEL' ,'TIJERA'], 1)) 

            # Imprimimos ambas elecciones para que el usuario lo pueda visualizar en cada jugada
            print(f'Has elegido {eleccion_jugador1}')
            print(f'Tu adversario ha elegido {eleccion_ordenador}')
            
            # Bloqueamos la jugada si el usuario hace una elección no válida (programación defensiva)
            if eleccion_jugador1 != 'PIEDRA' and eleccion_jugador1 != 'PAPEL' and eleccion_jugador1 != 'TIJERA':
                eleccion_jugador1 = input('Esa opción no es válida! Has de elegir piedra, papel o tijera').upper()

            # Definimos las condiciones para que puntúe un jugador u otro
            elif eleccion_jugador1 == 'PIEDRA' and eleccion_ordenador == 'TIJERA' or eleccion_jugador1 == 'PAPEL' and eleccion_ordenador == 'PIEDRA' or eleccion_jugador1 == 'TIJERA' and eleccion_ordenador == 'PAPEL':
                contador_jugador1 +=1
                print(f'¡Has ganado un punto!\nTus puntos totales son: {contador_jugador1} y los puntos de tu adversario: {contador_ordenador}\n' )
                    
            elif eleccion_jugador1 == 'PIEDRA' and eleccion_ordenador == 'PAPEL' or eleccion_jugador1 == 'PAPEL' and eleccion_ordenador == 'TIJERA' or eleccion_jugador1 == 'TIJERA' and eleccion_ordenador == 'PIEDRA':
                contador_ordenador +=1
                print(f'¡El ordenador ha ganado un punto!\nTus puntos totales son: {contador_jugador1} y los puntos de tu adversario: {contador_ordenador}\n')
                
            elif eleccion_jugador1 == 'PIEDRA' and eleccion_ordenador == 'PIEDRA' or eleccion_jugador1 == 'PAPEL' and eleccion_ordenador == 'PAPEL'or eleccion_jugador1 == 'TIJERA' and eleccion_ordenador == 'TIJERA':
                print(f'¡Habéis empatado!\nTus puntos totales son: {contador_jugador1} y los puntos de tu adversario: {contador_ordenador}\n')

            # Mostramos los mensajes de fin de juego cuando uno de los jugadores alcanza la puntuación requerida
            if contador_jugador1 == 3 or contador_ordenador == 3:
                if contador_jugador1 == 3:
                    print(f'¡El juego ha termninado!¡Has ganado! :)')
                elif contador_ordenador ==3:
                    print(f'¡El juego ha termninado! Has perdido :( Tu adversario ha ganado la partida')
    
    def ahorcado(self):

        #Creamos el dibujo asignando una variable a cada jugada perdida
        var1 = '==========='
        var2 = '         |\n         |\n         |\n         |\n         |\n==========='
        var3 = '   +-----+\n         |\n         |\n         |\n         |\n         |\n==========='
        var4 = '   +-----+\n   |     |\n         |\n         |\n         |\n         |\n==========='
        var5 = '   +-----+\n   |     |\n   O     |\n         |\n         |\n         |\n==========='
        var6 = '   +-----+\n   |     |\n   O     |\n   |     |\n         |\n         |\n==========='
        var7 = '   +-----+\n   |     |\n   O     |\n   |\    |\n         |\n         |\n==========='
        var8 = '   +-----+\n   |     |\n   O     |\n  /|\    |\n         |\n         |\n==========='
        var9 = '   +-----+\n   |     |\n   O     |\n  /|\    |\n    \    |\n         |\n==========='
        var10= '   +-----+\n   |     |\n   O     |\n  /|\    |\n  / \    |\n         |\n==========='

        dibujo =[var1,var2,var3,var4,var5,var6,var7,var8,var9,var10]

        #Creamos lista de palabras e importamos librería random
        lista_palabras = ["IGUANA", "PINTAR", "OBISPO", "PIÑATA", "OTORRINO", "MEMBRILLO", "REPASO"]
        secreta = random.choice(lista_palabras)


        #Creamos las variables que vamos a usar, y una función que muestre los guiones con el número de letras de la palabra
        secreta2 = secreta.upper()
        usadas = []
        secreta_guiones = []
        for letra in secreta:
            secreta_guiones.append('_')

        #Comienza el juego

        fallos = 0

        while fallos < 10 and '_' in secreta_guiones:
            if '_' in secreta_guiones:
                letra = input ('Dime una letra').upper()
                if letra not in usadas:
                    if letra in secreta:
                        print(f'¡Has acertado! La letra {letra} está en la palabra secreta:')
                        for indice,letras in enumerate (secreta):
                            if letras == letra:
                                secreta_guiones[indice] = letra
                        secreta = secreta.replace(letra,'-') 
                        print(''.join(secreta_guiones))
                        usadas.append(letra)
                    else:
                        print(f'La letra {letra} no está en la palabra secreta\n{dibujo[fallos]}')
                        fallos +=1
                        usadas.append(letra)
                else:
                    print(f'Ya has usado la letra {letra}')

        if '_' not in secreta_guiones:
                print(f'¡Has ganado! La palabra secreta es: {secreta2}')
        else:
            print(f'¡Lo sentimos!¡Has perdido la partida! La palabra secreta es: {secreta2}')

    def juego_preguntas_respuestas(self):
        
        #Creamos un diccionario con todas las preguntas
        diccionario_geo = {
            "Capital de España": "Madrid",
            "Río pasa por Madrid": "Manzanares",
            "¿En qué continente se encuentra Australia?": "Oceanía",
            "¿Qué país tiene más población?": "China",
            "¿Cuál es el océano más grande del mundo?": "Océano Pacífico",
            "¿Cuál es el río más largo del mundo?": "Amazonas",
            "¿Cuál es el punto más alto de la Tierra?": "Monte Everest",
            "¿En qué país se encuentra la Gran Muralla China?": "China",
            "¿Cuál es el país más grande del mundo en términos de superficie?": "Rusia",
            "¿Cuál es la capital de Francia?": "París",
            "¿Cuál es el punto más alto de sudámerica?" : "aconcagua",
            "¿En que país se encuentra el Taj mahal": "India",
            "¿Cuál es la capital de Italia?": "Roma",
            "¿Área fértil en el desierto?" : "Oasis",
            "¿Cuál es el lago más grande de África?": "Victoria"
        }

        #creamos una copia de la lista original y con shuffle se baraja la lista de preguntas
        preguntas_originales = list(diccionario_geo.keys())  
        preguntas_lista = list(preguntas_originales)  

        random.shuffle(preguntas_lista)  

        #Creamos contador del juego
        contador_aciertos = 0
        contador_fallos = 0

        #Comienza el juego
        while contador_aciertos < 5 and contador_fallos < 3:
            pregunta = preguntas_lista.pop() 
            respuesta = diccionario_geo[pregunta]
            respuesta_usuario = input(f'{pregunta} ')

            if respuesta_usuario == respuesta:
                contador_aciertos += 1
                print(f'¡Has acertado!\nTus aciertos son {contador_aciertos} y tus fallos son {contador_fallos}')
            else:
                contador_fallos += 1
                print(f'¡Has fallado!\nTus aciertos son {contador_aciertos} y tus fallos son {contador_fallos}')

        if contador_aciertos == 5:
            print('¡Has ganado! Enhorabuena :)')
        elif contador_fallos == 3:
            print('¡Has perdido! ¡Necesitas estudiar más! :(')
```
